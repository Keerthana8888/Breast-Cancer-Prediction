#Importing the Libraries

```python
# Import essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import Scikit-Learn libraries
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Import TensorFlow and Keras for building neural networks
import tensorflow as tf
from tensorflow import keras

# Install and import Gradio for creating a user interface for the model
!pip install gradio
import gradio as gr
```

#Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

#Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

#Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#Standardize the data

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

**Building the Neural Network**

download.png

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

##Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

##Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

##Gradio Interface

```python
def predict_tumor(
    mean_radius,
    mean_texture,
    mean_perimeter,
    mean_area,
    mean_smoothness,
    mean_compactness,
    mean_concavity,
    mean_concave_points,
    mean_symmetry,
    mean_fractal_dimension,
    radius_error,
    texture_error,
    perimeter_error,
    area_error,
    smoothness_error,
    compactness_error,
    concavity_error,
    concave_points_error,
    symmetry_error,
    fractal_dimension_error,
    worst_radius,
    worst_texture,
    worst_perimeter,
    worst_area,
    worst_smoothness,
    worst_compactness,
    worst_concavity,
    worst_concave_points,
    worst_symmetry,
    worst_fractal_dimension,
):
    input_data = (
        mean_radius,
        mean_texture,
        mean_perimeter,
        mean_area,
        mean_smoothness,
        mean_compactness,
        mean_concavity,
        mean_concave_points,
        mean_symmetry,
        mean_fractal_dimension,
        radius_error,
        texture_error,
        perimeter_error,
        area_error,
        smoothness_error,
        compactness_error,
        concavity_error,
        concave_points_error,
        symmetry_error,
        fractal_dimension_error,
        worst_radius,
        worst_texture,
        worst_perimeter,
        worst_area,
        worst_smoothness,
        worst_compactness,
        worst_concavity,
        worst_concave_points,
        worst_symmetry,
        worst_fractal_dimension,
    )

    input_data_as_numpy_array = np.asarray(input_data)
    input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
    input_data_std = scaler.transform(input_data_reshaped)
    prediction = model.predict(input_data_std)
    prediction_label = [np.argmax(prediction)]

    if prediction_label[0] == 0:
        return "The tumor is Malignant"
    else:
        return "The tumor is Benign"


# Create the Gradio interface
inputs = [
    gr.Number(label="Mean Radius"),
    gr.Number(label="Mean Texture"),
    gr.Number(label="Mean Perimeter"),
    gr.Number(label="Mean Area"),
    gr.Number(label="Mean Smoothness"),
    gr.Number(label="Mean Compactness"),
    gr.Number(label="Mean Concavity"),
    gr.Number(label="Mean Concave Points"),
    gr.Number(label="Mean Symmetry"),
    gr.Number(label="Mean Fractal Dimension"),
    gr.Number(label="Radius Error"),
    gr.Number(label="Texture Error"),
    gr.Number(label="Perimeter Error"),
    gr.Number(label="Area Error"),
    gr.Number(label="Smoothness Error"),
    gr.Number(label="Compactness Error"),
    gr.Number(label="Concavity Error"),
    gr.Number(label="Concave Points Error"),
    gr.Number(label="Symmetry Error"),
    gr.Number(label="Fractal Dimension Error"),
    gr.Number(label="Worst Radius"),
    gr.Number(label="Worst Texture"),
    gr.Number(label="Worst Perimeter"),
    gr.Number(label="Worst Area"),
    gr.Number(label="Worst Smoothness"),
    gr.Number(label="Worst Compactness"),
    gr.Number(label="Worst Concavity"),
    gr.Number(label="Worst Concave Points"),
    gr.Number(label="Worst Symmetry"),
    gr.Number(label="Worst Fractal Dimension"),
]

iface = gr.Interface(
    fn=predict_tumor,
    inputs=inputs,
    outputs="text",
    title="Breast Cancer Tumor Prediction",
    description="Enter the tumor characteristics to predict if it's benign or malignant.",
)

iface.launch()
```
